```python
import wikipedia
```

```python
def reading_material_summary(topic):

    material = wikipedia.summary(topic)

    print(material)
```

```python
def reading_material(topic):

    material = wikipedia.page(topic).content

    mat_end = material.find('== See also')
    if mat_end > 0:
        material = material[:mat_end]

    print(material)
```

```python
def create_reading_material(topic):

    material = wikipedia.page(topic).content

    mat_end = material.find('== See also')
    if mat_end > 0:
        material = material[:mat_end]

    f = open(dataset+'reading_material/'+topic+'.txt', 'w')
    f.write(material)
    f.close()
```

```python
create_reading_material('Microarchitecture')
```

```python
fails = []
for topic in topics:

    try:
        create_reading_material(topic)
    except:
        fails.append(topic)
        continue
```
