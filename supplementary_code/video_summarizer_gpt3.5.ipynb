```python
from openai import OpenAI

client = OpenAI(
    api_key=''
)
```

```python
client.models
```

```python
dataset = '../transcripts/'
text_data = dataset+'text/'
```

```python
import os
os
```

```python
filename = 'AI__transcript.txt'
```

```python
with open(text_data+filename) as f:
    text = f.readlines()
```

```python
text = text[0]
```

```python
completion = client.chat.completions.create(
  model="gpt-3.5-turbo-0125",
  messages=[
    {"role": "user", "content": "Summarize this transcript of an educational video into 500 words and include the important parts of the video: \n"+text},
  ]
)

summary = completion.choices[0].message.content
```
